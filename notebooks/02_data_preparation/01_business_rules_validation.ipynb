### Business Rule Validation
Validate dataset values against expected business rules and identify anomalies.

### 1 BUSINESS RULE CHECK

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
import pandas as pd

df = pd.read_csv('../../Data/processed/cleaned_customer_data_from_null_values.csv')
```

```python
# Age
df.loc[df["Age"] > 100, "Age"] = np.nan
df.loc[df["Age"] < 10, "Age"] = np.nan
```

```python
# Purchases
df.loc[df["Total_Purchases"] < 0, "Total_Purchases"] = np.nan
```

```python
# Credit balance
df.loc[df["Credit_Balance"] < 0, "Credit_Balance"] = np.nan
```

```python
# Lifetime value
df.loc[df["Lifetime_Value"] < 0, "Lifetime_Value"] = np.nan
```

### 2 RATE CHECK (0 - 100)

```python
rate_cols = [
    "Returns_Rate",
    "Discount_Usage_Rate",
    "Cart_Abandonment_Rate",
    "Email_Open_Rate"
]

for col in rate_cols:
    df.loc[df[col] > 100, col] = np.nan
    df.loc[df[col] < 0, col] = np.nan
```

### 3 NON-NEGATIVE FEATURES

```python
non_negative_cols = [
    "Login_Frequency",
    "Session_Duration_Avg",
    "Pages_Per_Session",
    "Wishlist_Items",
    "Customer_Service_Calls",
    "Product_Reviews_Written",
    "Mobile_App_Usage",
    "Social_Media_Engagement_Score"
]

for col in non_negative_cols:
    df.loc[df[col] < 0, col] = np.nan
```

### 4 RECENCY LIMIT (optional)

```python
df.loc[df["Days_Since_Last_Purchase"] > 365, "Days_Since_Last_Purchase"] = np.nan
```

### 5 IMPUTE MISSING (MEDIAN)

num_cols = df.select_dtypes(include="number").columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

### 6 FINAL DATA CHECK

```python
print("Remaining Missing Values:")
print(df.isna().sum().sort_values(ascending=False))

print("\nDataset Shape:", df.shape)

print("\nSummary Statistics:")
print(df.describe())
```

### 7 SAVE CLEAN DATA

```python
df.to_csv("../../Data/processed/cleaned_customer_data.csv", index=False)

print("\nCleaned dataset saved successfully.")
```
