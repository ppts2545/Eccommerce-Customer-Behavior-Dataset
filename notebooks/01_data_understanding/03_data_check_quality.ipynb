```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns   
import pandas as pd

df = pd.read_csv('../../Data/raw/ecommerce_customer_churn_dataset.csv')
```

```python
missing = df.isnull().sum()
missing_percent = (df.isnull().sum()/len(df))*100

missing_df = pd.DataFrame({
    "Missing Count": missing,
    "Missing %": missing_percent
}).sort_values("Missing %", ascending=False)

print("\nMissing Values in Each Column:")
missing_df
```

```python
print("\nDuplicate Rows in the Dataset:", df.duplicated().sum())
```

```python
df.dtypes
```

```python
df.describe()
```

```python
print("\nSkewness of Numeric Columns:")
df.skew(numeric_only=True)
```

Skew value	ความหมาย
    -0.5 ถึง 0.5	สมมาตร (normal ใกล้เคียง)

    0.5 ถึง 1	เอียงเล็กน้อย

    > 1	เอียงมาก

    < -1	เอียงมาก (เอียงซ้าย)

```python
sns.boxenplot(x=df["Membership_Years"])
plt.title("Boxen Plot of Membership Years")
plt.xlabel("Membership Years")
plt.ylabel("Frequency")
plt.show()

sns.histplot(df["Membership_Years"], kde=True, bins=30)
plt.title("Histogram of Membership Years")
plt.xlabel("Membership Years")
plt.ylabel("Frequency")
plt.show()

print("Minimum Membership Years:", df["Membership_Years"].min())
print("Maximum Membership Years:", df["Membership_Years"].max())
```
